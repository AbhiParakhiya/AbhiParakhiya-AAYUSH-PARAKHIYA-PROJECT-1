1. What is the difference between a function and a method in Python?give at least one example.

-> In Python, a function is a block of code that can be executed multiple times from different parts of your program. It's a self-contained piece of code that takes arguments, performs some operations, and returns a value. Functions are not part of a class or object; they are standalone.

-> In this example, greet is a function that takes a name argument and prints a greeting message. It's a standalone function that can be called multiple times with different arguments.

```
def greet(name: str) -> None:
    print(f"Hello, {name}!")

greet("John")  # Output: Hello, John!
greet("Alice")  # Output: Hello, Alice!
```

-> On the other hand, a method is a function that is part of a class or object. Methods are used to perform actions on an object or class, and they have access to the object's attributes and other methods. Methods are essentially functions that belong to a class or object.

```
class Person:
    def __init__(self, name: str, age: int) -> None:
        self.name = name
        self.age = age

    def greet(self) -> None:
        print(f"Hello, my name is {self.name} and I'm {self.age} years old!")

john = Person("John", 30)
john.greet()  # Output: Hello, my name is John and I'm 30 years old!

alice = Person("Alice", 25)
alice.greet()  # Output: Hello, my name is Alice and I'm 25 years old!
```

-> In this example, greet is a method of the Person class. It's a function that belongs to the class and has access to the object's attributes (self.name and self.age). The method is called on an instance of the class (john and alice), and it uses the object's attributes to print a personalized greeting message.

-> Notice how the method greet has a self parameter, which refers to the instance of the class. This is a key difference between functions and methods: methods always have a self parameter, which allows them to access and manipulate the object's attributes.

2. Explain the concept of function arguments and parameters in Python.give at least one example.

-> In Python, when we define a function, we specify parameters in the function definition. These parameters are like placeholders that will be replaced with actual values when the function is called.

-> On the other hand, when we call a function, we pass arguments to the function. These arguments are the actual values that are assigned to the parameters.

Here's an example to illustrate the difference:
Function definition with parameters:

```
def greet(name: str, age: int) -> None:
    print(f"Hello, {name}! You are {age} years old.")
```

In this example, name and age are parameters. They are placeholders that will be replaced with actual values when the function is called.

Function call with arguments:

```
greet("John", 30)
```
In this example, "John" and 30 are arguments. They are the actual values that are passed to the function and assigned to the parameters name and age, respectively.


3. What are the different ways to define and call a function in Python?give at least one example.

In Python, there are several ways to define and call a function. Here are some examples:

1. Simple Function Definition and Call

```
def greet(name: str) -> None:
    print(f"Hello, {name}!")
greet("John")
```    
Output: Hello, John!

In this example, we define a function greet with a single parameter name, and then call it by passing the argument "John".

2. Function with Default Argument Values

```
def greet(name: str, age: int = 25) -> None:
    print(f"Hello, {name}! You are {age} years old.")
greet("John")  # Output: Hello, John! You are 25 years old.
greet("Alice", 30)  # Output: Hello, Alice! You are 30 years old.    

```

In this example, we define a function greet with two parameters name and age, where age has a default value of 25. We can call the function with only one argument, and the default value will be used for age.

3. Function with Variable Number of Arguments (Varargs)

```
def greet(*names: str) -> None:
    for name in names:
        print(f"Hello, {name}!")
        
greet("John", "Alice", "Bob")
```
Output: Hello, John! Hello, Alice! Hello, Bob!

In this example, we define a function greet with a variable number of arguments *names, which allows us to pass any number of arguments to the function.

4. Function with Keyword Arguments

```
def greet(name: str, *, age: int, country: str) -> None:
    print(f"Hello, {name}! You are {age} years old from {country}.")
    
greet("John", age=30, country="USA")
```

Output: Hello, John! You are 30 years old from USA.

In this example, we define a function greet with keyword arguments age and country, which must be passed by name when calling the function.

5. Lambda Function

```
greet = lambda name: print(f"Hello, {name}!")
greet("John")  # Output: Hello, John!
```

In this example, we define a lambda function greet using the lambda keyword, which is a concise way to define small, one-line functions.

6. Nested Function

```
def outer():
    def inner(name: str) -> None:
        print(f"Hello, {name}!")
    inner("John")  # Output: Hello, John!

outer()
```
In this example, we define a nested function inner inside the outer function. The inner function is only accessible within the outer function.

These are just a few examples of the different ways to define and call functions in Python.

In this example, the greet function takes a name argument, constructs a greeting message, and returns it using the return statement. The returned value is then printed to the console.

5. What are iterators in Python and how do they differ from iterables?give at least one example.

In Python, iterators and iterables are two related but distinct concepts.

Iterables: An iterable is an object that can be iterated over, meaning it can be used in a for loop or with the iter() function to produce a sequence of values. Examples of iterables include lists, tuples, strings, dictionaries, and sets.

Iterators: An iterator is an object that keeps track of its position in a sequence and yields the next value in the sequence when its __next__() method is called. An iterator is created from an iterable using the iter() function.

Here's an example:
```
my_list = [1, 2, 3, 4, 5]  # iterable
my_iterator = iter(my_list)  # create an iterator from the iterable

print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
```

In this example, my_list is an iterable, and my_iterator is an iterator created from my_list. We can use the next() function to retrieve the next value from the iterator.

Notice that once an iterator is exhausted (i.e., all values have been retrieved), calling next() will raise a StopIteration exception.

Here's another example:
```
def infinite_sequence():
    n = 0
    while True:
        yield n
        n += 1

my_iterator = infinite_sequence()
print(next(my_iterator))  # Output: 0
print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
```
In this example, infinite_sequence is a generator function that yields an infinite sequence of numbers. We create an iterator from the generator function using the iter() function, and then use next() to retrieve values from the iterator.

Key differences between iterators and iterables:
An iterable is an object that can be iterated over, while an iterator is an object that keeps track of its position in a sequence.
An iterable can be used to create multiple iterators, while an iterator is specific to a single sequence.
An iterator can be exhausted, while an iterable can be iterated over multiple times.

6. Explain the concept of generators in Python and how they are defined.give at least one example.

In Python, a generator is a special type of iterable that generates a sequence of values on the fly, without storing them in memory all at once. Generators are defined using a generator function, which is a function that contains a yield statement.

Generator Function: A generator function is a function that uses the yield statement to produce a sequence of values. When a generator function is called, it returns a generator object, which is an iterator that produces the sequence of values.

Generator Object: A generator object is an iterator that is created by calling a generator function. It keeps track of its position in the sequence and yields the next value when its __next__() method is called.

Here's an example:


```
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

fib_gen = fibonacci(5)
print(next(fib_gen))  # Output: 0
print(next(fib_gen))  # Output: 1
print(next(fib_gen))  # Output: 1
print(next(fib_gen))  # Output: 2
print(next(fib_gen))  # Output: 3
```



In this example, the fibonacci function is a generator function that produces a sequence of Fibonacci numbers up to n. The yield statement is used to produce each value in the sequence. We create a generator object fib_gen by calling the fibonacci function, and then use the next() function to retrieve the next value from the generator.

Notice that the generator function only generates the next value in the sequence when next() is called, rather than storing the entire sequence in memory. This makes generators very memory-efficient, especially when working with large datasets.

* Key benefits of generators:

1. Memory efficiency: Generators only store the current state of the sequence, rather than the entire sequence.
2. Lazy evaluation: Generators only generate the next value in the sequence when needed, rather than computing the entire sequence upfront.
3. Flexibility: Generators can be used to generate infinite sequences, or sequences that depend on external factors.

7. What are the advantages of using generators over regular functions?give at least one example.

Advantages of using generators over regular functions:

1. Memory Efficiency: Generators use significantly less memory than regular functions, especially when dealing with large datasets. This is because generators only store the current state of the sequence, rather than the entire sequence.
2. Lazy Evaluation: Generators only generate the next value in the sequence when needed, rather than computing the entire sequence upfront. This can be particularly useful when working with expensive computations or I/O operations.
3. Flexibility: Generators can be used to generate infinite sequences, or sequences that depend on external factors.
Improved Performance: Generators can be faster than regular functions, especially when working with large datasets, since they avoid the overhead of creating and storing large lists.

Here's an example that demonstrates the memory efficiency of generators:


In this example, we define two functions: regular_function and generator_function. Both functions generate a sequence of squares of numbers from 0 to n-1. However, regular_function returns a list of all values, while generator_function returns a generator that yields each value one at a time.

8. What is a lambda function in Python and when is it typically used?give at least one example.

-> Lambda Functions in Python:

A lambda function, also known as an anonymous function, is a small, single-line function that can be defined inline within a larger expression. It's a shorthand way to create a function without declaring a separate named function using the def keyword.

The general syntax for a lambda function is:
```
lambda arguments: expression
```
Here, arguments is a comma-separated list of variables that will be passed to the function, and expression is the code that will be executed when the function is called.

-> When to use lambda functions:

Lambda functions are typically used in situations where a small, one-time-use function is needed, such as:

As an argument to a higher-order function (e.g., map(), filter(), reduce())
As a quick, one-off function for a specific task
When you need to create a function on the fly, without polluting the namespace with a named function

Example:

Let's say you want to create a function that squares a number. You can use a lambda function like this:
```
square = lambda x: x ** 2
print(square(5))  # Output: 25
```
In this example, the lambda function takes a single argument x and returns its square. You can then call the lambda function like a regular function, passing in the value 5 as an argument.

Another example is using a lambda function as an argument to the map() function:
```
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]
```

Here, the lambda function is used to square each element in the numbers list, and the resulting list is stored in squared_numbers.

Lambda functions are concise, expressive, and can be very useful in a variety of situations. However, they can also make the code harder to read if overused or used in complex scenarios, so use them judiciously!

10. What is the difference between 'map()`, reduce(), and `filter() functions in Python? give at least one example.

The map(), reduce(), and filter() Functions in Python:

These three functions are part of Python's functional programming toolkit, and they're often used together to process and transform data. Here's a brief overview of each function, along with examples:

1. map() Function:

As I mentioned earlier, the map() function applies a given function to each item of an iterable (such as a list, tuple, or string) and returns a list of the results.

Example:
```
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

```

2. filter() Function:
The filter() function applies a given function to each item of an iterable and returns a new iterable with only the items for which the function returns True.

Example:
```
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]

```
3. reduce() Function:
The reduce() function applies a rolling computation to sequential pairs of values in an iterable, reducing the iterable to a single output.

Note: In Python 3, reduce() is part of the functools module, so you need to import it explicitly: from functools import reduce.

Example:
```
from functools import reduce
import operator

numbers = [1, 2, 3, 4, 5]
product = reduce(operator.mul, numbers)
print(product)  # Output: 120
```

In this example, the reduce() function applies the operator.mul function (which multiplies two numbers) to the elements of the numbers list, effectively calculating the product of all elements.

Key differences:

1. map() transforms each element of an iterable, returning a new iterable with the same length.
2. filter() selects a subset of elements from an iterable, returning a new iterable with a potentially shorter length.
3. reduce() aggregates the elements of an iterable, returning a single output value.

These functions can be combined to perform complex data processing tasks. For example, you could use map() to transform data, filter() to select a subset of the data, and then reduce() to aggregate the results.

````python
def is_prime(n):
    if n <= 1:
        return print("Not Prime")
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return print("Not Prime")
    return print("Prime")
is_prime(101)
````

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

````python
def fibonacci_generator(num_terms):
    a, b = 0, 1
    count = 0
    while count < num_terms:
        yield a
        a, b = b, a + b
        count += 1

for number in fibonacci_generator(10):
    print(number)
````

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

````python
def powers_of_2(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

# Example usage:
for value in powers_of_2(5):
    print(value)
````

7. Implement a generator function that reads a file line by line and yields each line as a string.

````python
def read_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
````

10. Create a Python program that uses 'filter() to remove all the vowels from a given string.

````python
def remove_vowel(input_string):
  vowels = "AEIOUaieou"
  filtered_string = ''.join(filter(lambda x: x not in vowels, input_string))
  return filtered_string

remove_vowel("Abhi Prakhiya")
````
