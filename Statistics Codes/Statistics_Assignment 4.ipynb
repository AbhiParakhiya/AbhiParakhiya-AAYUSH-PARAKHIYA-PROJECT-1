1. What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It helps determine if there is enough evidence to support a certain claim or hypothesis about the data.

Key Concepts:
1. Null Hypothesis (ğ»0): Represents the default or no-effect assumption.
Example: "The new drug has no effect on blood pressure."

2. Alternative Hypothesis (ğ»1  or ğ»ğ‘): Represents the opposite of the null hypothesis, indicating a significant effect or difference.
Example: "The new drug reduces blood pressure."

3. Significance Level (ğ›¼):The threshold for rejecting ğ»0. Common values are 0.05 or 0.01, representing a 5% or 1% chance of rejecting ğ»0 when it is actually true.

4. p-value: The probability of obtaining the observed results (or more extreme) if ğ»0 is true.

5. Test Statistic: A value calculated from the sample data used to decide whether to reject  ğ»0.
 Examples include t-statistics, z-scores, and chi-square values.

3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean



```python
import numpy as np
import scipy.stats as stats

# Simulate a sample of 30 observations from a population with mean 50 and standard deviation 10
np.random.seed(42)
sample_size = 30
population_mean = 50
population_std = 10

# Generate a random sample from a normal distribution
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Sample mean
sample_mean = np.mean(sample_data)

# Z-test statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Print the results of the Z-test
print(f"Sample Mean: {sample_mean}")
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Decision rule (assuming alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot



```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate data: Assume the population mean is 50, and standard deviation is 10
np.random.seed(42)
sample_size = 30
population_mean = 50
population_std = 10

# Generate a random sample from a normal distribution
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Sample mean
sample_mean = np.mean(sample_data)

# Z-test statistic
z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# Print the results of the Z-test
print(f"Sample Mean: {sample_mean}")
print(f"Z-statistic: {z_statistic}")
print(f"P-value: {p_value}")

# Decision rule (assuming alpha = 0.05)
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha/2)  # Critical value for two-tailed test

# Print the decision
if abs(z_statistic) > critical_value:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")

# Visualization: Standard normal distribution and decision region
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(10, 6))

# Plot the standard normal curve
plt.plot(x, y, label="Standard Normal Distribution", color='blue')

# Shade the rejection regions (for two-tailed test)
plt.fill_between(x, y, where=(x <= -critical_value), color='red', alpha=0.5, label="Rejection Region (Left)")
plt.fill_between(x, y, where=(x >= critical_value), color='red', alpha=0.5, label="Rejection Region (Right)")

# Mark the critical values and the Z-statistic
plt.axvline(critical_value, color='black', linestyle='--', label=f"Critical Value = Â±{critical_value:.2f}")
plt.axvline(-critical_value, color='black', linestyle='--')
plt.axvline(z_statistic, color='green', linestyle=':', label=f"Z-statistic = {z_statistic:.2f}")

# Add labels and title
plt.title("Two-Tailed Z-Test and Decision Region")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type_errors(pop_mean, alt_mean, std_dev, sample_size, alpha=0.05):
    # Generate the population distribution under the null hypothesis (mean = pop_mean)
    null_distribution = norm(pop_mean, std_dev / np.sqrt(sample_size))
    
    # Generate the alternative hypothesis distribution (mean = alt_mean)
    alt_distribution = norm(alt_mean, std_dev / np.sqrt(sample_size))
    
    # Critical value for the rejection region based on alpha (two-tailed test)
    critical_value_upper = norm.ppf(1 - alpha/2, loc=pop_mean, scale=std_dev / np.sqrt(sample_size))
    critical_value_lower = norm.ppf(alpha/2, loc=pop_mean, scale=std_dev / np.sqrt(sample_size))
    
    # Plot the two distributions
    x = np.linspace(pop_mean - 4*std_dev, alt_mean + 4*std_dev, 1000)
    null_pdf = null_distribution.pdf(x)
    alt_pdf = alt_distribution.pdf(x)

    plt.figure(figsize=(10, 6))

    # Plot the null hypothesis distribution
    plt.plot(x, null_pdf, label="Null Hypothesis Distribution", color='blue')
    
    # Plot the alternative hypothesis distribution
    plt.plot(x, alt_pdf, label="Alternative Hypothesis Distribution", color='orange')
    
    # Shade the rejection regions (Type I Error)
    plt.fill_between(x, 0, null_pdf, where=(x <= critical_value_lower), color='red', alpha=0.5, label="Type I Error Region (Left)")
    plt.fill_between(x, 0, null_pdf, where=(x >= critical_value_upper), color='red', alpha=0.5, label="Type I Error Region (Right)")
    
    # Shade the Type II error region (for the alternative hypothesis)
    plt.fill_between(x, 0, alt_pdf, where=(x >= critical_value_lower) & (x <= critical_value_upper), color='green', alpha=0.5, label="Type II Error Region")

    # Mark the critical values
    plt.axvline(critical_value_upper, color='black', linestyle='--', label=f"Critical Value (Upper) = {critical_value_upper:.2f}")
    plt.axvline(critical_value_lower, color='black', linestyle='--', label=f"Critical Value (Lower) = {critical_value_lower:.2f}")
    
    # Mark the means
    plt.axvline(pop_mean, color='blue', linestyle='-', label=f"Population Mean = {pop_mean}")
    plt.axvline(alt_mean, color='orange', linestyle='-', label=f"Alternative Mean = {alt_mean}")
    
    # Title and Labels
    plt.title("Visualization of Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    
    # Show the plot
    plt.show()

# Parameters
pop_mean = 50        # Population mean (Null Hypothesis)
alt_mean = 52        # Alternative mean (when the null is false)
std_dev = 10         # Population standard deviation
sample_size = 30     # Sample size
alpha = 0.05         # Significance level

# Call the function to visualize Type I and Type II errors
visualize_type_errors(pop_mean, alt_mean, std_dev, sample_size, alpha)
```

7. Perform a paired sample T-test using Python and visualize the comparison results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate data: pre-test and post-test scores for 30 participants
np.random.seed(42)
pre_test = np.random.normal(loc=50, scale=10, size=30)  # pre-test scores
post_test = pre_test + np.random.normal(loc=5, scale=5, size=30)  # post-test scores with a slight increase

# Perform a paired sample T-test
t_statistic, p_value = stats.ttest_rel(pre_test, post_test)

# Print the results of the T-test
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Decision rule (assuming alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between pre-test and post-test scores.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between pre-test and post-test scores.")

# Visualize the results with a boxplot and paired differences
# Create a figure with two subplots: one for the pre-test vs post-test scores and one for the differences
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plotting the pre-test and post-test scores
axes[0].boxplot([pre_test, post_test], vert=True, patch_artist=True, labels=["Pre-test", "Post-test"])
axes[0].set_title("Pre-test vs Post-test Scores")
axes[0].set_ylabel("Scores")

# Calculate and plot the differences
differences = post_test - pre_test
axes[1].hist(differences, bins=10, color='skyblue', edgecolor='black')
axes[1].set_title("Differences (Post-test - Pre-test)")
axes[1].set_xlabel("Difference")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

8. Simulate data and perform both Z-test and T-test, then compare the results using Python.



```python
import numpy as np
import scipy.stats as stats

# Set seed for reproducibility
np.random.seed(42)

# Simulate two random samples (sample sizes n1 = 50, n2 = 50)
# Assume both samples are from normal distributions with means of 10 and 12, and standard deviations of 5
sample1 = np.random.normal(loc=10, scale=5, size=50)
sample2 = np.random.normal(loc=12, scale=5, size=50)

# Sample sizes
n1, n2 = len(sample1), len(sample2)

# Sample means
mean1, mean2 = np.mean(sample1), np.mean(sample2)

# Population standard deviation (for Z-test)
population_std = 5  # Known population standard deviation

# Z-test: Two-sample Z-test (Assume known population standard deviation)
z_statistic = (mean1 - mean2) / (population_std * np.sqrt(1/n1 + 1/n2))
# P-value for two-tailed test
z_p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

# T-test: Two-sample T-test (Assume unknown population standard deviation)
t_statistic, t_p_value = stats.ttest_ind(sample1, sample2)

# Print the results of both tests
print(f"Z-test statistic: {z_statistic}")
print(f"Z-test p-value: {z_p_value}")
print(f"T-test statistic: {t_statistic}")
print(f"T-test p-value: {t_p_value}")

# Decision rule (assuming alpha = 0.05)
alpha = 0.05
print("\nDecision based on Z-test:")
if z_p_value < alpha:
    print("Reject the null hypothesis: The means are significantly different (based on Z-test).")
else:
    print("Fail to reject the null hypothesis: The means are not significantly different (based on Z-test).")

print("\nDecision based on T-test:")
if t_p_value < alpha:
    print("Reject the null hypothesis: The means are significantly different (based on T-test).")
else:
    print("Fail to reject the null hypothesis: The means are not significantly different (based on T-test).")
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance


```python
import scipy.stats as stats
import numpy as np

def confidence_interval(sample_data, confidence_level=0.95):
    # Calculate the sample mean and standard error
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # sample standard deviation
    sample_size = len(sample_data)
    
    # Z-score for the given confidence level
    z_score = stats.norm.ppf((1 + confidence_level) / 2)
    
    # Calculate the margin of error
    margin_of_error = z_score * (sample_std / np.sqrt(sample_size))
    
    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return (lower_bound, upper_bound)

# Example usage
sample_data = [23, 25, 30, 28, 27, 29, 31, 24, 26, 25]
conf_interval = confidence_interval(sample_data)
print(f"Confidence Interval: {conf_interval}")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.



```python
def margin_of_error(sample_data, confidence_level=0.95):
    # Calculate the sample standard deviation
    sample_std = np.std(sample_data, ddof=1)  # Using ddof=1 for sample standard deviation
    sample_size = len(sample_data)
    
    # Z-score for the given confidence level
    z_score = stats.norm.ppf((1 + confidence_level) / 2)
    
    # Calculate the margin of error
    margin_error = z_score * (sample_std / np.sqrt(sample_size))
    
    return margin_error

# Example usage
sample_data = [23, 25, 30, 28, 27, 29, 31, 24, 26, 25]
confidence_level = 0.95  # For 95% confidence level
error = margin_of_error(sample_data, confidence_level)
print(f"Margin of Error: {error}")
```

```python
def bayesian_inference(prior_H, likelihood_E_given_H, likelihood_E_given_not_H):
    # P(H|E) = (P(E|H) * P(H)) / P(E)
    # P(E) = P(E|H)*P(H) + P(E|~H)*P(~H)
    prior_not_H = 1 - prior_H  # Prior probability of not having the disease
    
    # Calculate P(E) (the total probability of the evidence)
    likelihood_E = (likelihood_E_given_H * prior_H) + (likelihood_E_given_not_H * prior_not_H)
    
    # Calculate P(H|E), the posterior probability of having the disease given a positive test result
    posterior_H_given_E = (likelihood_E_given_H * prior_H) / likelihood_E
    
    return posterior_H_given_E

# Example values
prior_H = 0.01  # Prior probability of having the disease (1%)
likelihood_E_given_H = 0.95  # Probability of testing positive given disease (sensitivity)
likelihood_E_given_not_H = 0.05  # Probability of testing positive without disease (false positive rate)

# Compute the posterior probability
posterior = bayesian_inference(prior_H, likelihood_E_given_H, likelihood_E_given_not_H)
print(f"The probability of having the disease given a positive test result is {posterior:.4f}")
```

```python
import numpy as np

def calculate_expected_frequencies(observed_data):
    # Calculate the row totals
    row_totals = observed_data.sum(axis=1, keepdims=True)
    
    # Calculate the column totals
    column_totals = observed_data.sum(axis=0, keepdims=True)
    
    # Calculate the total sample size (sum of all observed values)
    total_sum = observed_data.sum()
    
    # Calculate the expected frequencies
    expected_frequencies = (row_totals @ column_totals) / total_sum
    
    return expected_frequencies

# Example observed data (contingency table: 2x3)
observed_data = np.array([[30, 10, 20],  # Male
                          [20, 25, 15]]) # Female

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Display expected frequencies
print("Observed Data: \n", observed_data)
print("\nExpected Frequencies: \n", expected_frequencies)
```

16. Implement an F-test using Python to compare the variances of two random samples.



```python
import numpy as np
import scipy.stats as stats

def f_test(sample1, sample2):
    # Calculate the variances of the two samples
    var1 = np.var(sample1, ddof=1)  # ddof=1 to get sample variance
    var2 = np.var(sample2, ddof=1)  # ddof=1 to get sample variance
    
    # Calculate the F-statistic (larger variance / smaller variance)
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1
    
    # Degrees of freedom for the two samples
    df1 = len(sample1) - 1  # Degrees of freedom for sample1
    df2 = len(sample2) - 1  # Degrees of freedom for sample2
    
    # Calculate the p-value using the F-distribution
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))
    
    return f_statistic, p_value

# Example data: two random samples with different variances
sample1 = np.random.normal(loc=10, scale=5, size=100)  # Sample 1 (variance = 25)
sample2 = np.random.normal(loc=10, scale=3, size=100)  # Sample 2 (variance = 9)

# Perform the F-test
f_stat, p_value = f_test(sample1, sample2)

# Display the results
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the results
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The variances are different.")
else:
    print("\nFail to reject the null hypothesis: The variances are the same.")
```


19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def check_normality(group):
    """ Check if a group is normally distributed using the Shapiro-Wilk test. """
    stat, p_value = stats.shapiro(group)
    if p_value > 0.05:
        return True  # The group is normally distributed
    else:
        return False  # The group is not normally distributed

def check_equal_variance(*groups):
    """ Check if variances are equal using Levene's test. """
    stat, p_value = stats.levene(*groups)
    if p_value > 0.05:
        return True  # Variances are equal
    else:
        return False  # Variances are not equal

def plot_data(groups):
    """ Plot histograms and boxplots to visually check assumptions. """
    # Histogram to check normality
    plt.figure(figsize=(12, 6))
    for i, group in enumerate(groups, start=1):
        plt.subplot(1, len(groups), i)
        sns.histplot(group, kde=True)
        plt.title(f'Group {i} Histogram')
    plt.tight_layout()
    plt.show()

    # Boxplot to check equal variance
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=groups, palette="Set3")
    plt.title("Boxplot of Groups")
    plt.xlabel("Group")
    plt.ylabel("Values")
    plt.show()

def check_assumptions(*groups):
    """ Check the assumptions of normality, equal variance, and independence. """
    print("\nChecking assumptions for ANOVA...\n")

    # 1. Check normality for each group
    normality_results = {}
    for i, group in enumerate(groups, start=1):
        normality_results[f'Group {i} Normality'] = check_normality(group)

    # 2. Check equal variance (homoscedasticity)
    equal_variance = check_equal_variance(*groups)

    # Print results
    for group, result in normality_results.items():
        print(f"{group}: {'Normal' if result else 'Not normal'}")
    
    print(f"\nEqual variance assumption: {'Met' if equal_variance else 'Not met'}")
    
    # Plotting the data for visual checks
    plot_data(groups)

# Example data: Three groups of data
group1 = np.random.normal(loc=23, scale=5, size=30)
group2 = np.random.normal(loc=19, scale=6, size=30)
group3 = np.random.normal(loc=25, scale=5, size=30)

# Check the assumptions for ANOVA
check_assumptions(group1, group2, group3)
```

```python
import numpy as np
import scipy.stats as stats

# Simulate a sample from a normal distribution
np.random.seed(42)

# Sample data (e.g., heights, weights, etc.)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10, Sample Size=30

# Hypothesized population variance (sigma_0^2)
sigma_0_squared = 100  # Assuming population variance is 100 (sigma = 10)

# Step 1: Calculate sample variance
sample_variance = np.var(sample_data, ddof=1)  # Sample variance (ddof=1 for sample variance)

# Step 2: Calculate the test statistic
n = len(sample_data)  # Sample size
chi_square_statistic = (n - 1) * sample_variance / sigma_0_squared

# Step 3: Calculate the critical value and p-value for the Chi-square test
alpha = 0.05  # Significance level (5%)

# Degrees of freedom
df = n - 1

# Critical values for two-tailed test
critical_value_lower = stats.chi2.ppf(alpha / 2, df)
critical_value_upper = stats.chi2.ppf(1 - alpha / 2, df)

# Step 4: Decision Rule and Interpretation
print(f"Sample Variance: {sample_variance}")
print(f"Chi-Square Statistic: {chi_square_statistic}")
print(f"Critical Value Lower: {critical_value_lower}")
print(f"Critical Value Upper: {critical_value_upper}")

# Test the hypothesis
if chi_square_statistic < critical_value_lower or chi_square_statistic > critical_value_upper:
    print("Reject the null hypothesis: The population variance is significantly different from the hypothesized variance.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the sample variance and the hypothesized population variance.")
```


25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.


```python
import numpy as np
import scipy.stats as stats

# Sample data
# Group 1: 100 trials with 45 successes
# Group 2: 120 trials with 55 successes
x1, n1 = 45, 100  # Successes and sample size for Group 1
x2, n2 = 55, 120  # Successes and sample size for Group 2

# Step 1: Calculate proportions
p1 = x1 / n1  # Proportion of success in Group 1
p2 = x2 / n2  # Proportion of success in Group 2

# Step 2: Calculate the pooled proportion
p = (x1 + x2) / (n1 + n2)

# Step 3: Calculate the Z-statistic
z = (p1 - p2) / np.sqrt(p * (1 - p) * (1 / n1 + 1 / n2))

# Step 4: Calculate the p-value for the two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Step 5: Interpretation
alpha = 0.05  # Significance level

print(f"Z-statistic: {z}")
print(f"P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```
