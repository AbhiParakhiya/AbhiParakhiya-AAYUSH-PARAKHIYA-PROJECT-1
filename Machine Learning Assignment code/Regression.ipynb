1. What is Simple Linear Regression?



--> Simple Linear Regression is the relationship between one Independent variable and One Dependent Variable using a straight line.


2. What are the key assumptions of Simple Linear Regression?


--> 1. Linearity

2. Independence

3. Homoscedasticity (constant variance)

4. Normality of errors

5. No outliers with severe influence


3. What does the coefficient m represent in the equation Y=mX+c?



--> It indicates how much Y changes when X increases by 1 unit.

4. What does the intercept c represent in the equation Y=mx+c?



--> It is the value of Y when X = 0.

5. How do we calculate the slope m in Simple Linear Regression?



--> m=∑(x−xˉ)(y−yˉ)/∑(x−xˉ)^2
	​


6. What is the purpose of the least squares method in Simple Linear Regression?



--> To minimize the squared difference between observed and predicted values.

7. How is the coefficient of determination (R) interpreted in Simple Linear Regression?



--> It measures the percentage of variation in Y explained by X.

Example:
R² = 0.80 → 80% of variability is explained.

8. What is Multiple Linear Regression?



--> Multiple Linear Regression is the relationship between one Independent variable and Many Dependent Variable using a multi-dimension plane.

9. What is the main difference between Simple and Multiple Linear Regression?



--> Main Difference is 

SLR --> it has one independent variable and one dependent variable, one straight line, Easier.

MLR --> it has one independent variable and many dependent variable, multidimensional plane, complex.

10. What are the key assumptions of Multiple Linear Regression?



--> 1. Linearity

2. Independence

3. Homoscedasticity

4. Normality

5. No multicollinearity

11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?



--> When error variance is not constant across X values.

Effect: Unreliable confidence intervals and hypothesis tests.

12. How can you improve a Multiple Linear Regression model with high multicollinearity?



--> 1. Remove correlated variables

2. PCA

3. Regularization (Ridge / Lasso)

4. VIF analysis

13. What are some common techniques for transforming categorical variables for use in regression models?



--> 1. One-Hot Encoding

2. Label Encoding

3. Target Encoding

14. What is the role of interaction terms in Multiple Linear Regression?



--> They model combined effects of variables.

Example:
Price = b1Area + b2Age + b3*(Area×Age)

15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?



--> Simple: starting Y value

Multiple: expected Y when ALL X = 0

16. What is the significance of the slope in regression analysis, and how does it affect predictions?



--> Determines direction and strength of relationship.

17. How does the intercept in a regression model provide context for the relationship between variables?


--> Baseline reference point for predictions.


18. What are the limitations of using R' as a sole measure of model performance?


--> Doesn’t detect overfitting

Always increases with more variables

Misleading without context
Use Adjusted R² instead.


19. How would you interpret a large standard error for a regression coefficient?


--> Unstable coefficient

Low confidence in estimate

Possibly weak predictors


20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?


--> Funnel pattern in residual plot

Breusch-Pagan test


21. What does it mean if a Multiple Linear Regression model has a high Re but low adjusted RP?



--> Model fits training data well but includes weak variable

22. Why is it important to scale variables in Multiple Linear Regression?


-->Faster convergence

Fair coefficient comparison

Helps regularization
