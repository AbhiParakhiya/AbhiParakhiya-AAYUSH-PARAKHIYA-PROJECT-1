2. What is correlation?

--> Correlation measures the strength and direction of the relationship between two variables. It ranges from -1 to 1. A value near 1 means strong positive relationship, near -1 indicates a strong negative one.

3. What does negative correlation mean?

--> A negative correlation implies that as one variable increases, the other tends to decrease. For example, more hours spent watching TV may negatively correlate with academic performance.

4. Define Machine Learning. What are the main components in Machine Learning?

--> Machine Learning is a method where machines learn from data to make predictions. Main components include:
Data
Model
Loss function
Optimization algorithm
Evaluation metrics

5. How does loss value help in determining whether the model is good or not?

--> The loss value quantifies how well a model's predictions match actual outcomes. A lower loss value means better performance. During training, minimizing this loss leads to improved model accuracy.

6. What are continuous and categorical variables?

--> Continuous variables are numeric and can take any value (e.g., height, temperature).

Categorical variables represent discrete labels or classes (e.g., gender, color, city names).

8. What do you mean by training and testing a dataset?

--> Training dataset: Used to fit and train the model.

Testing dataset: Used to evaluate how well the model performs on unseen data. It helps in detecting overfitting.

9. What is sklearn.preprocessing?

--> It is a module in Scikit-learn that offers functions for data preprocessing. It includes tools for scaling, encoding, imputing missing values, and transforming features to prepare them for model training.

10. What is a Test set?

--> A test set is a portion of the dataset used only to assess the performance of a trained model. It is kept separate from the training data to ensure unbiased evaluation.
