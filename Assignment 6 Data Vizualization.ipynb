### 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

#### np.random.seed(20)

#### data = {

#### 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'],

#### 'Day': np.random.choice(range (1, 31), 100),100),

#### 'Sales': np.random.randint(1000, 5000, 100)

#### } 

#### df = pd.DataFrame (data)

```python
np.random.seed(20)
data = {'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'],100), 'Day': np.random.choice(range (1, 31), 100), 'Sales': np.random.randint(1000, 5000, 100)} 
df = pd.DataFrame(data)
df
```

```python
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')


heatmap_data = heatmap_data.reset_index()


heatmap_long = heatmap_data.melt(id_vars='Day', var_name='Month', value_name='Sales')


fig = px.density_heatmap(
    heatmap_long,
    x='Month',
    y='Day',
    z='Sales',
    color_continuous_scale='turbo',
    title='Heatmap of Sales Across Months and Days'
)

fig.update_layout(
    xaxis_title="Month",
    yaxis_title="Day",
    coloraxis_colorbar=dict(title="Sales")
)

# Show the plot
fig.show()
```

### 1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python

from bokeh.plotting import figure, show,output_file
import numpy as np

# Generate x and y data for the sine wave
x = np.linspace(0, 10, 500)  # x values from 0 to 10
y = np.sin(x)  # y values as sine of x

# Create a Bokeh figure
p = figure(
    title="Sine Wave Plot",
    x_axis_label="x",
    y_axis_label="sin(x)",
    width=800,
    height=400
)

# Add a sine wave line to the figure
p.line(x, y, line_width=2, color="blue", legend_label="sin(x)")

# Customize the legend
p.legend.location = "top_left"
p.legend.title = "Function"

# Show the plot
show(p)

```

### 2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
from bokeh.models import ColumnDataSource


# Generate random data
np.random.seed(42)  # For reproducibility
x = np.random.rand(100) * 10  # x values between 0 and 10
y = np.random.rand(100) * 10  # y values between 0 and 10
sizes = np.random.rand(100) * 50 + 10  # Sizes between 10 and 60
colors = np.random.choice(["red", "blue", "green", "orange", "purple"], size=100)  # Random colors

# Create a ColumnDataSource
source = ColumnDataSource(data={
    "x": x,
    "y": y,
    "sizes": sizes,
    "colors": colors,
})


p = figure(
    title="Scatter Plot with Variable Sizes and Colors",
    x_axis_label="x",
    y_axis_label="y",
    width=800,
    height=400,
)

# Add scatter points to the figure
p.scatter(
    x="x",
    y="y",
    size="sizes",
    color="colors",
    source=source,
    alpha=0.7,
    legend_field="colors",  # Legend based on colors
)

# Customize the legend
p.legend.title = "Colors"
p.legend.location = "top_right"

# Show the plot
show(p)
```

### 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

#### fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data={
    "fruits": fruits,
    "counts": counts,
})

# Create a Bokeh figure
p = figure(
    x_range=fruits,  # Set x-axis to categorical range
    title="Fruit Counts",
    x_axis_label="Fruits",
    y_axis_label="Counts",
    width=800,
    height=400,
)

# Add a bar glyph to the plot
p.vbar(
    x="fruits", 
    top="counts", 
    width=0.5, 
    source=source, 
    color="dodgerblue", 
    legend_label="Counts"
)

# Customize the plot
p.legend.title = "Legend"
p.legend.location = "top_right"

# Show the plot
show(p)
```

### 4. Create a Bokeh histogram to visualize the distribution of the given data.

#### data_hist = np.random.randn(1000) hist, edges np.histogram(data_hist, bins=30)

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(
    title="Histogram of Random Data",
    x_axis_label="Value",
    y_axis_label="Frequency",
    width=800,
    height=400,
)

# Add a quad glyph for the histogram
p.quad(
    top=hist, 
    bottom=0, 
    left=edges[:-1], 
    right=edges[1:], 
    fill_color="dodgerblue", 
    line_color="white", 
    alpha=0.7,
    legend_label="Frequency"
)

# Customize plot appearance
p.legend.title = "Legend"
p.legend.location = "top_right"

# Show the plot
show(p)
```

### 5. Create a Bokeh heatmap using the provided dataset.

#### data_heatmap = np.random.rand(10, 10) x = np.linspace(0, 1, 10) y = np.linspace(0, 1, 10) xx, yy np.meshgrid(x, y)

```python
from bokeh.plotting import figure, show
from bokeh.io import output_file
import numpy as np
from bokeh.palettes import Viridis256
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.transform import linear_cmap

# Generate data
data_heatmap = np.random.rand(10, 10)  # 10x10 random data
x = np.linspace(0, 1, 10)  # x-values
y = np.linspace(0, 1, 10)  # y-values
xx, yy = np.meshgrid(x, y)  # Create a meshgrid for coordinates

# Flatten the grid and the data to make them compatible for plotting
x_coords = xx.flatten()
y_coords = yy.flatten()
values = data_heatmap.flatten()

# Create a ColumnDataSource from the data
source = ColumnDataSource(data=dict(x=x_coords, y=y_coords, value=values))

# Create a Bokeh figure
p = figure(
    title="Heatmap",
    x_axis_label="X",
    y_axis_label="Y",
    width=600,
    height=600
)

# Set the color mapper using linear_cmap
mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(values), high=max(values))

# Plot the heatmap using rect glyphs with color mapping
p.rect(
    x='x', 
    y='y', 
    width=0.1, 
    height=0.1, 
    source=source,  # Use the ColumnDataSource here
    color=mapper,   # Use the color mapper here
    line_color="white",  # Line color between cells
    legend_label="Heatmap",
    alpha=0.6
)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
output_file("heatmap.html")
show(p)
```
