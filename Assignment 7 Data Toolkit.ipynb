### 3. Explain the following terms:
#### a) The difference in np.array, np.asarray and np.asanyarray. 




The three NumPy functions—np.array, np.asarray, and np.asanyarray—are used to create arrays, but they differ in their behavior, especially in handling input data types and subclasses. Here's an explanation of the differences:

1. np.array()
Purpose: Creates a new NumPy array object, regardless of whether the input is already a NumPy array.
Key Characteristics:
Always returns a new array unless copy=False is explicitly set, and the input is already a NumPy array.
If the input is an array-like object (e.g., Python lists, tuples), it converts it into a NumPy array.
Default behavior copies the data.
Does not preserve subclasses like matrix.

```python
import numpy as np

# Example
list_data = [1, 2, 3]
array1 = np.array(list_data)  # Creates a new NumPy array

existing_array = np.array([4, 5, 6])
array2 = np.array(existing_array)  # Creates a copy by default

print("Original Array:", array1)
print("Copy of Existing Array:", array2)
```

2. np.asarray()
Purpose: Converts an input to a NumPy array, but avoids copying the input if it is already a NumPy array.
Key Characteristics:
If the input is a NumPy array of the same type, no copy is made; the same object is returned.
If the input is not a NumPy array, it is converted into one.
Does not preserve subclasses like matrix.



```python
list_data = [1, 2, 3]
array1 = np.asarray(list_data)  # Converts to NumPy array

existing_array = np.array([4, 5, 6])
array2 = np.asarray(existing_array)  # Does not create a copy

print("Converted Array:", array1)
print("Same Array without Copy:", array2, id(array2) == id(existing_array))  # True

```

3. np.asanyarray()
Purpose: Similar to np.asarray(), but it preserves subclasses of NumPy arrays (e.g., np.matrix or user-defined subclasses).
Key Characteristics:
Converts input to a NumPy array unless the input is already an array or an array subclass.
If the input is an array subclass, the same subclass is preserved.
Useful when working with custom subclasses of arrays.
python

```python
matrix_data = np.matrix([[1, 2], [3, 4]])
array1 = np.asanyarray(matrix_data)  
array2 = np.asarray(matrix_data)    

print("Preserved Subclass (matrix):", array1)
print("Converted to ndarray:", array2)
```

#### b) The difference between Deep copy and shallow copy.

1. Shallow Copy
Definition: A shallow copy creates a new object but does not create copies of the objects that the original object references. Instead, it only copies references to those objects.
Behavior:
The outer object is a new object.
The inner objects (nested or referenced objects) are not duplicated; they are shared between the original and the copy.
Effect:
Modifying inner objects will reflect in both the original and the shallow copy.
Modifying the outer object does not affect the original.

```python
import copy

# Original object
original = [[1, 2, 3], [4, 5, 6]]

# Shallow copy
shallow = copy.copy(original)

# Modify shallow copy
shallow[0][0] = 99

print("Original:", original)
print("Shallow Copy:", shallow)
```

2. Deep Copy
Definition: A deep copy creates a new object and recursively copies all objects referenced by the original.
Behavior:
The outer object is a new object.
All inner (nested) objects are also recursively duplicated.
Effect:
Modifying any object in the copy does not affect the original, and vice versa.

```python
import copy

# Original object
original = [[1, 2, 3], [4, 5, 6]]

# Deep copy
deep = copy.deepcopy(original)

# Modify deep copy
deep[0][0] = 99

print("Original:", original)
print("Deep Copy:", deep)


```

### 15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:

#### a) Plot a histogram of the data with 30 bins.

#### b) Overlay a line plot representing the normal distribution's probability density function (PDF).

#### c) Label the x-axis as 'Value and the y-axis as 'Frequency/Probability.

#### d) Set the title of the plot as 'Histogram with PDF Overlay. I


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data
data = np.random.normal(1,10, size=1000)

# Create histogram
bins = np.linspace(min(data), max(data), 30)
count, bins, _ = plt.hist(data, bins=bins, density=True, alpha=0.6, color='teal', label='Histogram')

# Calculate bin centers
bin_centers = 0.5 * (bins[1:] + bins[:-1])

# Fit the data and compute the PDF
mean, std_dev = norm.fit(data)
pdf = norm.pdf(bin_centers, loc=mean, scale=std_dev)

# Plot the PDF
plt.plot(bin_centers, pdf, 'r-', lw=2, label=f'PDF (Mean={mean:.2f}, Std={std_dev:.2f})')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Add legend
plt.legend()

# Display the plot
plt.show()
```

### 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot.

```python
np.random.seed(42)
x = np.random.uniform(-10, 10, 100)  # Random x values
y = np.random.uniform(-10, 10, 100)  # Random y values
x
```

```python
y
```

```python
def get_quadrant(x, y):
    if x > 0 and y > 0:
        return "Q1"
    elif x < 0 and y > 0:
        return "Q2"
    elif x < 0 and y < 0:
        return "Q3"
    elif x > 0 and y < 0:
        return "Q4"
    else:
        return "Origin"
quadrants = [get_quadrant(x_val, y_val) for x_val, y_val in zip(x, y)]
```

```python
data = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Create the scatter plot
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='Set2', s=100, edgecolor='k')

# Add lines for axes
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)

# Customize the plot
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Quadrant-wise Scatter Plot")
plt.legend(title="Quadrant", loc="upper left")
plt.show()
```

### 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

```python
from bokeh.plotting import figure, show
import numpy as np

# Generate data for the sine wave
x = np.linspace(0, 10, 500)  # x values from 0 to 10
y = np.sin(x)                # y values as sine of x

# Create a Bokeh figure
p = figure(
    title="Sine Wave Function",       # Set the title
    x_axis_label="x-axis",           # Label for the x-axis
    y_axis_label="y-axis",           # Label for the y-axis
    width=800,                       # Plot width
    height=400                       # Plot height
)

# Add a line plot for the sine wave
p.line(x, y, line_width=2, color="blue", legend_label="sin(x)")

# Customize grid lines
p.xgrid.grid_line_color = "gray"
p.ygrid.grid_line_color = "gray"
p.xgrid.grid_line_dash = "dashed"
p.ygrid.grid_line_dash = "dashed"

# Add legend and styling
p.legend.location = "top_left"
p.legend.label_text_font_size = "10pt"

# Display the plot
show(p)
```

### 19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart.

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import random

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = [random.randint(10, 100) for _ in categories]

# Prepare the data source
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a Bokeh figure
p = figure(
    x_range=categories,                   # Set x-axis as categorical
    title="Random Categorical Bar Chart", # Title
    x_axis_label="Categories",            # Label for x-axis
    y_axis_label="Values",                # Label for y-axis
    width=800, height=400                 # Figure size
)

# Add a color mapper for the bars
color_mapper = linear_cmap(field_name="values", palette=Viridis256, low=min(values), high=max(values))

# Add bars to the plot
p.vbar(
    x='categories', 
    top='values', 
    width=0.5, 
    source=source, 
    color=color_mapper, 
    line_color="black"
)

# Add hover tool for tooltips
hover = HoverTool(
    tooltips=[
        ("Category", "@categories"),
        ("Value", "@values")
    ]
)
p.add_tools(hover)

# Customize grid and axes
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```
